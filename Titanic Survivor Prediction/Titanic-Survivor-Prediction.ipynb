### Model 1 </br>
Columns excluded - PassengerId, Name, Ticket, Cabin

```python
df.columns
```

```python
x = df.iloc[:,[2,4,5,6,7,9,11]]
y = df.iloc[:,1]
print(x.shape,y.shape)
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

Model fitting is a measure of how well a machine learning model generalizes to similar data to that on which it was trained.</br> A model that is well-fitted produces more accurate outcomes.</br> A model that is overfitted matches the data too closely.</br> A model that is underfitted doesn't match closely enough.

```python
def model_insights(model):
  m = model.fit(x_train,y_train)
  ypred = m.predict(x_test)
  print('Confusion Matrix\n',confusion_matrix(y_test,ypred))
  print('Accuracy Score\n',accuracy_score(y_test,ypred)) #Tests accuracy
  print('Classification Report\n',classification_report(y_test,ypred))
```

# KNN

```python
m1 = KNeighborsClassifier(n_neighbors=21)
model_insights(m1)
```

# LogRegg

```python
m2 = LogisticRegression()
model_insights(m2)
```

# Decision Tree

```python
m3 = DecisionTreeClassifier(criterion = 'gini',max_depth = 12)
model_insights(m3)
```

# SVM

```python
m4 = SVC(kernel = 'linear',C = 1.0)
model_insights(m4)
```

# Random Forest Classifier

```python
m5 = RandomForestClassifier(n_estimators=30,criterion='gini',max_depth=10)
model_insights(m5)
```
